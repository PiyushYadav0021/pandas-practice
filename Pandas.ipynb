# Merging , Joining & Concatenation

```python
#Let's take Data frame

Teachers = pd.DataFrame({
    'teachers_id': [1, 2, 3, 4, 5],
    'Name': ['Dilip lode', 'Satish muthmare', 'Rajender londe', 'Sunil tope', 'Dipika Padhukone'],
    'Subjects': ['Hindi', 'Marathi', 'English', 'Science', 'Biology']
})

# DataFrame 2: Salary information
salary = pd.DataFrame({
    'teachers_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
Teachers
```

```python
salary
```

```python
# Merging the dataframes

pd.merge(Teachers, salary, on ='teachers_id')
```

```python
# There are some other functions like inner, outer, left, right to merge the dataframes


pd.merge(Teachers, salary, on ='teachers_id', how = 'inner')
```

inner use to match the common of teacher and salary write only that

```python
pd.merge(Teachers, salary, on ='teachers_id', how = 'outer')
```

Outer use to write all the rows doesn't matter match or not, it write it with NaN

```python
pd.merge(Teachers, salary, on ='teachers_id', how = 'left')
```

left is use to write all columns of Teachers

```python
pd.merge(Teachers, salary, on ='teachers_id', how = 'right')
```

Right is use to write all columns of salary

# Group by Aggregation

```python
Stores = {
    'Francisy' : ['F1','F2','F2','F1','F1','F1','F2','F2'],
    'Things' : ['Tea','Coffee','Tea','Coffee','Coffee','Tea','Coffee','Tea'],
    'quantity' : [3,4,2,6,1,7,8,5],
    'Sales' : [20,30,33,43,12,11,56,87],
    'Dates' : pd.date_range('27-07-2025', periods = 8)
}
```

```python
DF = pd.DataFrame(Stores)
DF
```

```python
DF.groupby('Francisy')

#groupby doesn't create any dataframe just create a object that we can't see that object
```

```python
See = DF.groupby('Francisy')
See
#We have created See to check the group we can't
```

```python
# To see the group we use for loop

for i, n in See:
    print(i)
    print(n)
```

```python
# To find the sum of sales

DF.groupby('Francisy')['Sales'].sum()
```

```python
#Group of multiple rows

DF.groupby(['Francisy','Things'])['Sales'].sum()
```

# Pivot Table

```python
Data = {
    'Date': pd.date_range('27-07-2025', periods =30),
    'Product' : ['A','B','C','D','E','F'] * 5,
    'Region' : ['East', 'West', 'North', 'South','East', 'West', 'North', 'South','East', 'West', 'North', 'South','East', 'West', 'North', 'South','East', 'West', 'North', 'South','East', 'West', 'North', 'South','East', 'West', 'North', 'South','East', 'West'],
    'Sales' : np.random.randint(10, 1000, 30),
    'Units' : np.random.randint(10, 1000, 30),
    'People' : ['Dilip lode', 'Satish muthmare', 'Rajender londe', 'Sunil tope', 'Dipika Padhukone', 'Dilip lode', 'Satish muthmare', 'Rajender londe', 'Sunil tope', 'Dipika Padhukone','Dilip lode', 'Satish muthmare', 'Rajender londe', 'Sunil tope', 'Dipika Padhukone','Dilip lode', 'Satish muthmare', 'Rajender londe', 'Sunil tope', 'Dipika Padhukone','Dilip lode', 'Satish muthmare', 'Rajender londe', 'Sunil tope', 'Dipika Padhukone','Dilip lode', 'Satish muthmare', 'Rajender londe', 'Sunil tope', 'Dipika Padhukone']
}
```

```python
DF = pd.DataFrame(Data)

DF['Month'] = DF['Date'].dt.month_name()
DF['Quarter'] = 'Q' + DF['Date'].dt.quarter.astype(str)
DF
```

```python
pd.pivot_table(DF, values = "Sales", index = "Region", columns = "Product")
```

The Values in this are mean
we can find min, max , median, std also. in next we will find median

```python
pd.pivot_table(DF, values ="Sales", index = "Region", columns = "Product", aggfunc = "median")
```

Now, this is a median of these values

```python
pd.pivot_table(DF, values = ["Sales","Units"], index = "Region", columns = ["Product","Quarter"] )
```
